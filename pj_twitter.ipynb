```python
# making a new column for age group
import numpy as np
tweet_grp=['0-500.000','500.001-1000.000','1.000.001-1.500.000','1.500.001-2.000.000','2.000.001-2.500.000','2.500.001-3.000.000']
tw['tweet_count_group'] = pd.cut(tw.tweet_count,bins=np.arange(0,3000001,500000),labels=tweet_grp,right=True)
tw.head()
```

```python
tw = tw.drop(['gender_gold','profile_yn_gold','tweet_coord','tweet_id','tweet_location','user_timezone', '_unit_state', 'created', 'tweet_created', 'sidebar_color', 'link_color', 'profileimage', '_last_judgment_at', '_trusted_judgments'],axis = 1)
tw.head()
```

```python
## Heating map
import seaborn as sns
sns.heatmap(tw.corr(),annot=True,fmt='.1g',cbar=False)
```

```python
## Finding the User IDs which have tweet_count greater than 200000
tw = tw[tw['tweet_count'] >= 200000]
tw
```

```python
## Computing the z-scores
tw['z_tweet_count'] = (tw['tweet_count'] - tw['tweet_count'].mean()) / tw['tweet_count'].std()
tw['z_retweet_count'] = (tw['retweet_count'] - tw['retweet_count'].mean()) / tw['retweet_count'].std()
tw.head()
```

```python
from sklearn.cluster import KMeans

## Defining the k-means model
three_clusters = KMeans(n_clusters = 11, n_init = 20)

## applying k-means to data
tw_clusters = three_clusters.fit(tw[['z_tweet_count']])

## Extracting and appending the cluster labels
tw['cluster'] = tw_clusters.labels_
tw
```

```python
## Hirachical clustering
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc

plt.figure(figsize = (10, 8))
dend = shc.dendrogram(shc.linkage(tw[['z_tweet_count']], method = 'average'))
plt.show()
```

```python
import statsmodels.formula.api as smf

## Building the linear model
lm_md = smf.ols(formula = 'tweet_count ~ retweet_count', data = tw).fit()

## Extracting model result
lm_md.summary()
```
